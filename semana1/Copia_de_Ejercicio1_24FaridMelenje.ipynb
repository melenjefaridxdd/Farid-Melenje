# Taller básico pandas - numpy

```python
import pandas as pd
```

```python
# Crear un dataframe que tenga 3 registros(filas) cada una debe contener el nombre de una cancion, el artista que la canta, y la duracion en segundos
NombreCancion = (['Ferxxo30','Doblexxo','Bubalu'])
Artista = (['Ferxxo','Ferxxo feat Jbalvin','Ferxxo feat Rema'])
Duracionsegundos = ([173,210,300])

df_canciones = pd.DataFrame({'Canción': NombreCancion, 'Artist': Artista, 'Segundos':Duracionsegundos})
df_canciones
```

```python
# Imprimir la cola del dataframe
df_canciones.tail()
```

```python
# Acceder a la columna duracion
df_canciones['Segundos']
```

```python
#Calcular el promedio de las duraciones
df_canciones.describe()
promedio = df_canciones['Segundos'].mean()
print(promedio)
```

```python

# Obtener los nombres de las canciones que tengan duracion mayor que el promedio
df_canciones[df_canciones['Segundos'] > promedio]
```

```python
# Agregar una columna al dataframe que contenga True si la duracion es mayor que 3 minutos
# y false si es menor
df_canciones  = df_canciones.assign(Mayor3minutos = df_canciones['Segundos'] > 180)
df_canciones
```

```python
# Convertir el dataframe a numpy array
arraycanciones = df_canciones.to_numpy()
arraycanciones
```

```python
# Imprimir el shape del array
arraycanciones.shape
```

```python
# Hacer un reshape para convertir el arreglo de 3x3 a un arreglo de 12 elementos
arraycanciones2 = arraycanciones.reshape(12)
arraycanciones2
```
